```python
import cv2
import os
from ultralytics import YOLO

# Load the YOLOv8 model
model = YOLO("D:/PaperCYolov8s.1a/PaperCYolov8s.1/runs/detect/train/weights/best.pt")

# Constants for distance calculation
ACTUAL_HEIGHT = 0.5  # meters, replace with the real-world height of your object
FOCAL_LENGTH = 500   # pixels, calculated from camera calibration

# Function to estimate distance based on bounding box height
def estimate_distance(bbox_height, actual_height=ACTUAL_HEIGHT, focal_length=FOCAL_LENGTH):
    return (actual_height * focal_length) / bbox_height

# Access the webcam (change to 0 if the camera index is different)
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()  # Read the webcam frame
    if not ret:
        break

    # Perform inference on the current frame
    results = model.predict(frame, conf=0.5, iou=0.4)

    # Extract detection results for the first image in the results
    detections = results[0].boxes

    # Iterate over detected objects
    for detection in detections:
        # Bounding box coordinates (x1, y1, x2, y2) and confidence
        x1, y1, x2, y2 = map(int, detection.xyxy[0])  # Extract bounding box coordinates
        bbox_height = y2 - y1  # Calculate the height of the bounding box

        # Estimate distance based on the bounding box height
        distance = estimate_distance(bbox_height)

        # Draw the bounding box and distance text on the frame
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(frame, f"Distance: {distance:.2f}m", (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

    # Display the frame with annotations
    cv2.imshow("Webcam YOLOv8", frame)

    # Exit on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        cap.release()  # Release the webcam
        cv2.destroyAllWindows()  # Close OpenCV windows
        os._exit(0)  # Stop the kernel immediately
```

```python
import cv2
from ultralytics import YOLO

# Load the YOLOv8 model
model = YOLO("D:/PaperCYolov8s.1a/PaperCYolov8s.1/runs/detect/train/weights/best.pt")

# Access the webcam
cap = cv2.VideoCapture(0)

# Define class groups
group_a_classes = [1, 3, 5, 7]  # Class IDs for Group A
group_b_classes = [0, 2, 4, 6]  # Class IDs for Group B

# Loop to continuously get frames from the webcam
while True:
    ret, frame = cap.read()  # Capture a frame
    if not ret:
        print("Failed to capture frame. Exiting...")
        break

    # Perform inference with YOLOv8
    results = model.predict(frame, conf=0.5, iou=0.4)

    # Create lists to store new group labels and colors
    grouped_labels = []
    box_colors = []  # Initialize box_colors for each frame

    for detection in results[0].boxes:
        if int(detection.cls[0]) in group_a_classes:
            grouped_labels.append("Recyclable")  # Label for Group A
            box_colors.append((0, 255, 0))  # Green color for Group A
        elif int(detection.cls[0]) in group_b_classes:
            grouped_labels.append("Non-recyclable")  # Label for Group B
            box_colors.append((0, 0, 255))  # Red color for Group B
        else:
            grouped_labels.append("Unknown")  # If it doesn't match any group
            box_colors.append((255, 255, 255))  # White color for unknown

    # Draw bounding boxes with new group labels and colors
    for i, box in enumerate(results[0].boxes.xyxy):
        x1, y1, x2, y2 = map(int, box)  # Coordinates of the bounding box
        label = grouped_labels[i]
        color = box_colors[i]  # Get color based on group

        # Draw rectangle and label on the frame
        cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
        cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    # Display the annotated frame
    cv2.imshow("Webcam YOLOv8 with Grouped Classes", frame)

    # Exit the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources after the loop ends
cap.release()
cv2.destroyAllWindows()
```

```python
import torch
import cv2
import numpy as np
from ultralytics import YOLO
from pytorch_grad_cam import GradCAM
from pytorch_grad_cam.utils.image import show_cam_on_image

# Load the YOLO model
model = YOLO("D:/PaperCYolov8s.1a/PaperCYolov8s.1/runs/detect/train/weights/best.pt")

# Load an image
img_path = r"D:\TestImages&videos\Single\Single (1).jpg"
img = cv2.imread(img_path)

# Ensure the image is loaded
if img is None:
    raise FileNotFoundError(f"Image not found at {img_path}")

# Convert BGR to RGB and normalize
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) / 255.0
img_resized = cv2.resize(img_rgb, (640, 640))  # Resize to YOLO input size

# Convert image to tensor and ensure it tracks gradients
img_tensor = torch.tensor(img_resized).permute(2, 0, 1).unsqueeze(0).float()
img_tensor = img_tensor.requires_grad_()  # Ensure gradients are tracked

# Move tensor to GPU if available
device = "cuda" if torch.cuda.is_available() else "cpu"
img_tensor = img_tensor.to(device)
model.model = model.model.to(device)

# Select a target layer (e.g., a convolutional layer near the output)
target_layer = model.model.model[-3]  # Typically one of the last Conv layers

# Prepare Grad-CAM
class YOLOWrapper(torch.nn.Module):
    """Wrapper to make YOLO compatible with Grad-CAM."""
    def __init__(self, yolo_model):
        super(YOLOWrapper, self).__init__()
        self.yolo_model = yolo_model

    def forward(self, x):
        outputs = self.yolo_model(x)  # Perform forward pass
        return outputs[0]  # Return only feature maps for Grad-CAM

# Wrap the YOLO model
wrapped_model = YOLOWrapper(model.model)

# Create Grad-CAM object
cam = GradCAM(model=wrapped_model, target_layers=[target_layer], reshape_transform=None)

# Perform Grad-CAM
with torch.no_grad():
    grayscale_cam = cam(input_tensor=img_tensor)  # No targets needed for general visualization

# Convert Grad-CAM output to overlay visualization
visualization = show_cam_on_image(img_resized, grayscale_cam[0], use_rgb=True)

# Show the visualization
cv2.imshow("Grad-CAM", visualization)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import torch
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Load the YOLOv8 model
model = YOLO("D:/PaperCYolov8s.1a/PaperCYolov8s.1/runs/detect/train/weights/best.pt")

# Define a list to store features
features = []

# Hook function to capture the output of specific layers
def hook_fn(module, input, output):
    features.append(output)

# Register hooks for Conv2D layers
for name, layer in model.model.named_modules():
    if isinstance(layer, torch.nn.Conv2d):  # Target Conv2D layers
        layer.register_forward_hook(hook_fn)

# Load an example image
image_path = r"D:\TestImages&videos\Single\single (28).jpg"

# Preprocess and pass the image through the model
results = model(image_path)  # Trigger the hooks

# Check if features are captured
if features:
    print(f"Captured {len(features)} feature maps.")
else:
    print("No features captured. Check your hook registration.")

# Visualize features from the first captured layer
if features:
    feature_map = features[0].detach().cpu().numpy()

    # Plot the first 4 channels
    for i in range(4):
        plt.subplot(1, 4, i + 1)
        plt.imshow(feature_map[0, i, :, :], cmap='gray')
        plt.axis('off')
    plt.show()
```

```python
# Visualize edge detection (features from the first convolutional layer)
edge_feature_map = features[0].detach().cpu().numpy()

# Plot the first 4 channels
plt.figure(figsize=(10, 5))
for i in range(4):
    plt.subplot(1, 4, i + 1)
    plt.imshow(edge_feature_map[0, i, :, :], cmap='gray')
    plt.title(f'Edge {i}')
    plt.axis('off')
plt.show()
```

```python
import cv2

# Load the original image
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Plot histograms for R, G, B channels
colors = ('r', 'g', 'b')
plt.figure(figsize=(10, 5))
for i, color in enumerate(colors):
    hist = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.plot(hist, color=color)
    plt.title("Color Histogram")
    plt.xlim([0, 256])
plt.show()
```

```python
# Visualize texture features (features from an intermediate layer)
texture_feature_map = features[2].detach().cpu().numpy()  # Adjust the index for deeper layers

plt.figure(figsize=(10, 5))
for i in range(4):  # Display the first 4 channels
    plt.subplot(1, 4, i + 1)
    plt.imshow(texture_feature_map[0, i, :, :], cmap='viridis')
    plt.title(f'Texture {i}')
    plt.axis('off')
plt.show()
```
